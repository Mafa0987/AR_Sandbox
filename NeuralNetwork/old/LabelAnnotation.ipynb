```python
import cv2

x1 = 0
y1 = 0 
number = 0
path = "images/PalmColor/"

# function which will be called on mouse input
def drawRectangle(action, x, y, flags, *userdata):
  global x1
  global y1
  global number
  if action == cv2.EVENT_LBUTTONDOWN:
    x1 = x
    y1 = y
    print("x1: " + str(x1) + " y1: " + str(y1) + " number: " + str(number))
    f = open("images/PalmPositions/" + str(number) + ".txt", "x")
    f.write(str(x1) + " " + str(y1))
    f.close()
    number += 1
    if number > 100:
      cv2.destroyAllWindows()
    else:
      image = cv2.imread(path + str(number) + ".jpg")
      cv2.imshow("Window", image)


cv2.namedWindow("Window")
cv2.setMouseCallback("Window", drawRectangle)

image = cv2.imread("images/PalmColor/" + str(number) + ".jpg")
cv2.imshow("Window", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt

number = 499
im =  Image.open("Data/PNG_A/Images/ClosedHand/" + str(number) + ".png")
im = im.resize((240,240), Image.LANCZOS)
f = open("Data/PNG_A/Positions/ClosedHand/" + str(number) + ".txt", "r")
x = f.read().split()
f.close()
x1 = int(x[0])/300*240
y1 = int(x[1])/400*240
im = np.array(im)
for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        if (i-y1)**2 + (j-x1)**2 < 10:
            im[i][j] = [255,255,255]

plt.figure()
plt.imshow(im)
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt
import cv2

number = 0
cv2.namedWindow("Window")


while(True):
    im =  Image.open("images/PalmColor/" + str(number) + ".jpg")
    im = im.resize((240,240), Image.LANCZOS)
    f = open("images/PalmPositions/" + str(number) + ".txt", "r")
    x = f.read().split()
    f.close()
    x1 = int(x[0])/300*240
    y1 = int(x[1])/400*240
    im = np.array(im)
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            if (i-y1)**2 + (j-x1)**2 < 10:
                im[i][j] = [255,255,255]
    cv2.imshow("Window", im)
    cv2.waitKey(0)
    if number == 100:
        break
    number += 1


cv2.destroyAllWindows()
```
