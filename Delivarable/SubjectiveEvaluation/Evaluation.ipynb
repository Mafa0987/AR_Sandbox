```python
import pandas as pd
import numpy as np

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('data.csv')

# Display the first few rows of the DataFrame to verify the contents
print(df)
```

```python
def get_mean_and_variance(data, columns_):
    results = np.array([])
    for column_ in columns_:
        print(data.columns[column_] + '\n')
        results = np.append(results, data[data.columns[column_]].values)
    mean = results.mean()
    variance = results.var()
    standard_deviation = results.std()
    return mean, variance, standard_deviation
```

```python
#Precense and Engagement
Precensemean, Precensevariance, Precensestandard_deviation = get_mean_and_variance(df, [10, 11])

print('Presenece and Engagement')
print('Mean: ' + str(Precensemean))
print('Variance: ' + str(Precensevariance))
print('Standard Deviation: ' + str(Precensestandard_deviation))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the variables (replace with actual data)
means = [Precensemean, Immersionmean, Flowmean, Usabilitymean, Emotionmean, Skillmean, Attractivenessmean, Techmean]
std_devs = [Precensestandard_deviation, Immersionstandard_deviation, Flowstandard_deviation, Usabilitystandard_deviation, Emotionstandard_deviation, Skillstandard_deviation, Attractivenessstandard_deviation, Techstandard_deviation]
labels = ['Precense', 'Immersion', 'Flow', 'Usability', 'Emotion', 'Skill', 'Attractiveness', 'Technology Adaption']

# Plot the means with standard deviation
plt.figure(figsize=(10, 4))
x = np.arange(len(labels))

# Create bar plot
bars = plt.bar(x, means, yerr=std_devs, capsize=5, color='skyblue', edgecolor='black')

# Customize the plot
plt.xlabel('Variables')
plt.ylabel('Mean')
plt.title('Means with Standard Deviation')
plt.xticks(x, labels, rotation=45)
plt.ylim(0, 6)

# Add value labels on top of the bars
for bar, std in zip(bars, std_devs):
    yval = bar.get_height()
    plt.text(bar.get_x() -0.25 + bar.get_width()/2.0, yval, f'{round(yval, 2)}\n±{round(std, 2)}', va='bottom', ha='center') # va: vertical alignment, ha: horizontal alignment

plt.show()
```

```python
import numpy as np

# Sample data
age = df['Age'].values

# Create a DataFrame
data_age = pd.DataFrame({'Age': age})

# Count occurrences of each age
age_counts = data_age['Age'].value_counts()

# Reindexing to include all ages between 20 to 35
all_ages = np.arange(20, 36)
age_counts = age_counts.reindex(all_ages, fill_value=0)

# Plotting Age
plt.figure(figsize=(10, 6))
ax = age_counts.plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.ylim(0, 5.5)
plt.xticks(rotation=0)

# Adding percentage labels
total = age_counts.sum()
for i, count in enumerate(age_counts):
    percentage = (count / total) * 100
    plt.text(i, count + 0.1, f'{percentage:.2f}%', ha='center')

plt.show()

```
